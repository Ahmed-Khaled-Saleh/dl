## How to Run

- Install the library (I use nbdev for development, nbdev works by making the code as a library).
- `main.py` in the root directory is the main script to run the project.
- A st of cfgs are available under the `cfgs/` directory.
- A set of job script files (Mahti) is available under the `jobs/` directory.
- Supporting files from kaggle and the dataset should be placed under the `data/` directory.
- Models are available under the `eval/models/` directory.
